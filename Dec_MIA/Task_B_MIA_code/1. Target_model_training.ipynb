```python
import pandas as pd
from simpletransformers.classification import ClassificationModel,ClassificationArgs

train_df = pd.read_pickle("data/train_olid_pickle")
test_df = pd.read_pickle("data/test_olid_pickle")


model_args = ClassificationArgs()
model_args.num_train_epochs = 1
model_args.output_dir = 'models/target_model'
model_args.manual_seed = 2412
model_args.overwrite_output_dir = True

model = ClassificationModel("bert", "google-bert/bert-base-cased", args=model_args,use_cuda=True)
model.train_model(train_df)


result, model_outputs, wrong_predictions = model.eval_model(test_df)
result
```
