```python
import pandas as pd

splits = {'train': 'train.csv', 'test': 'test.csv'}
train_df = pd.read_csv("hf://datasets/christophsonntag/OLID/" + splits["train"])
test_df = pd.read_csv("hf://datasets/christophsonntag/OLID/" + splits["test"])

train_df = train_df[['tweet','subtask_a']].rename(columns = {"tweet":"text","subtask_a":"labels"})
test_df = test_df[['tweet','subtask_a']].rename(columns = {"tweet":"text","subtask_a":"labels"})

def label_mapping(labels):
    """
    map OFF to 1 and NOT to 0
    """
    bin_label = None
    if labels == "OFF":
        bin_label = 1
    if labels == "NOT":
        bin_label = 0
    return bin_label

train_df['labels'] = train_df['labels'].apply(label_mapping)
test_df['labels'] = test_df['labels'].apply(label_mapping)

train_df.to_pickle('data/train_olid_pickle')
test_df.to_pickle('data/test_olid_pickle')
```
