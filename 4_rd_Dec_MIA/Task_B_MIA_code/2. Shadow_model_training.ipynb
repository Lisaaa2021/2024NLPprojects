```python
import pandas as pd
from simpletransformers.classification import ClassificationModel,ClassificationArgs

member_olid = pd.read_pickle("data/train_olid_pickle")[:5000] #member 5k,
member_hasoc = pd.read_csv("data_shadow/hasoc_train.csv")[:3000][['text','labels']] #member 3k


nonmember_olid = pd.read_pickle("data/train_olid_pickle")[5000:10000] #non-member 5k
nonmember_hasoc = pd.read_csv("data_shadow/hasoc_train.csv")[3000:][['text','labels']] #non-member #2852

train_member = pd.concat([member_olid,member_hasoc]).sample(frac = 1, random_state = 1207)
```

```python
model_args = ClassificationArgs()
model_args.num_train_epochs = 1
model_args.output_dir = 'models/shadow_model'
model_args.manual_seed = 174124
model_args.overwrite_output_dir = True

model = ClassificationModel("bert", "google-bert/bert-base-cased", args=model_args,use_cuda=True)
model.train_model(train_member)
```
